## Identify Skewness

### Normal Distribution

We can see industries that follow a normal distrubution of salary if their box plot is symmetric. Industries in Transportation Equipment Manuf., IT Services, Internet, Enterprise, and Computer Hardware & Software industries follow this characteristic which reveals that most software engineers in these industries fall in the middle to make average salary in their respective range, minus the few outliers. 



----------
### Positive Skew 

We can see industries that have an unysmmetric box plot where the upper quartile is slightly larger to be positively skewed. The Video Game and Electrical & Electronic Manufacturing industries show that people in these industries have reported to have a larger salary than the mean salary in their respective industry.
